```
# Evaluate the total Asset of the current coorperating company
import pandas as pd
import numpy as np
data = pd.read_csv("Data/partners.csv")
```

```
df = data['Total Assets'].astype(str)
print(df)

df['totalAssetMil'] = df.str.extract('\$(\d+)\,').fillna(0).astype(int)
df['totalAssetThou'] = df.str.extract('\,(\d+)\,').fillna(0).astype(int)
df['Y0M'] = df[['totalAssetThou','totalAssetThou']].dot([100,1])
```

```
df.head()
```

```
#get rid of the "$"
def remove_punctuation(string):
    new_string = ""
    for i in string:
        if i != "$":
            new_string += i
    return new_string
```

```
new_df = []
for dollars in df:
    a = remove_punctuation(dollars)
    if dollars != "nan":
        new_df.append(a.replace(",",""))




print(new_df)
```

```
#get the median of the total Assets

for i in range(len(new_df)):
    new_df[i] = int(new_df[i])
new_df.sort()
print(new_df[9])
```
