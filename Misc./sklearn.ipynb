```
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.naive_bayes import MultinomialNB
```

```
#only need mission statement and whether or not they partner together
current_part = pd.read_csv("current_partners.csv")
current_part["Partners?"] = 1
current_df = current_part.iloc[:,1:3]
current_df = current_df.rename(columns = {"Mission Statement":"Mission"})
print(current_df)
```

```
new_part = pd.read_csv("Ecosystem.csv")
new_part["Partners?"] = 0
new_df = new_part[['Mission', 'Partners?']].copy()
print(new_df)
```

```
partner_df = pd.concat([current_df, new_df], axis=0)
partner_df.dropna(how='any', inplace=True)
print(partner_df)
```

```
x = partner_df["Mission"]
y = partner_df["Partners?"]
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.30, random_state = 53)
print(x_train.shape, x_test.shape, y_train.shape, y_test.shape)
```

```
# counts the words in a text ==> turning into numerical data that algorithm understands
count_vectorizer = CountVectorizer(stop_words = "english")
# turns into a matrix with the training and test data
count_train = count_vectorizer.fit_transform(x_train.values)
count_test = count_vectorizer.transform(x_test.values)
```

```
nb_classifier = MultinomialNB()

# fit the classifier to the training data
nb_classifier.fit(count_train, y_train)
```

```
pred = nb_classifier.predict(count_test)

#check how accurate the data was sorted/ classified
score = metrics.accuracy_score(y_test, pred)
print(score)
```

```
#confusion matrix -> what was predicted correctly and what wasn't
cm = metrics.confusion_matrix(y_test, pred, labels= [1, 0])
print(cm)
```
