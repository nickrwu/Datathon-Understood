```
import pandas as pd
import numpy as np
data = pd.read_csv("Data/partners.csv")
df = data['Total Revenue'].astype(str).tolist()
print(df)
```

```
#get rid of the "$"
def remove_punctuation(string):
    new_string = ""
    for i in string:
        if i != "$":
            new_string += i
    return new_string
```

```
new_df = []
for dollars in df:
    a = remove_punctuation(dollars)
    if dollars != "nan":
        new_df.append(a.replace(",",""))
```

```
#get the mean/median of the total assets
for i in range(len(new_df)):
    new_df[i] = int(new_df[i])
new_df.sort()
print(new_df[9])
```
