```
import pandas as pd
```

```
data = pd.read_csv("current_partners.csv")
```

```
common_words = {}
```

```
new_data = data["Mission Statement"]
print(new_data)
type(new_data)
```

```
# creating a list from the columns
new_data_ls = new_data.to_list()
print(new_data_ls)
```

```
data_list = [] # new list where theres a sublist of words in each mission statement
for sentence in new_data_ls:
    data_list.append(sentence.split(" "))

print(data_list)
```

```
for block in data_list:
    for word in block:
        if word in common_words:
            common_words[word] += 1
        else:
            common_words[word] = 1
```

```
new_dict = {}
for word in common_words:
    if len(word) > 4 and common_words[word] > 1:
        new_dict[word] = common_words[word]

sorted(new_dict.items(), key=lambda x: x[1], reverse = True)
```
